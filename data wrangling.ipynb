2.Data Quality Assessment :

```python

print(df.isnull().sum())
print(df.duplicated().sum())
print(df['Region'].unique())
print(df['Product'].unique())
import matplotlib.pyplot as plt
df.boxplot(column='Sales')
plt.title("Sales Outliers")
plt.show()
```

## Data Quality Assessment

- The dataset contains no missing values.
- No duplicate records were found.
- The 'Date' column is stored as an object data type and needs to be converted to datetime format.
- Unique values in categorical columns ('Region', 'Product') were checked and found to be consistent.
- A boxplot was used to detect outliers in the 'Sales' column, and no significant outliers were observed.

## Conclusion

The dataset is clean with no missing values or duplicates. The categorical data is consistent, and no major outliers were detected. However, the 'Date' column requires conversion to datetime format for proper analysis.

3.Data Cleaning & Transformation :

```python

df['Date'] = pd.to_datetime(df['Date'])
df['Region'] = df['Region'].str.strip().str.title()
df['Product'] = df['Product'].str.strip().str.title()
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df = df.sort_values(by='Date')
df.reset_index(drop=True, inplace=True)
df.to_csv("cleaned_sales_data.csv", index=False)
df.head()
```

## Data Cleaning & Transformation

- The 'Date' column was converted to datetime format.
- Text columns such as 'Region' and 'Product' were standardized.
- New features were created, including 'Month' and 'Year' from the 'Date' column.
- The dataset was sorted by date and the index was reset.

## Conclusion

The dataset has been successfully cleaned and transformed. Data quality issues were addressed, and new features were created to enhance analysis. The dataset is now ready for further analysis.
